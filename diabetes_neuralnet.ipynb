# <font color ='orange'> Diabetes Neural Network for Diagnosing Women

### <font color ='blue'>Values in the datasets represent the following columns
Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age,Outcome

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
np.random.seed(7)
```

```python
dataset=np.loadtxt('prima-indians-diabetes.csv',delimiter=',')

x=dataset[:,0:8]
y=dataset[:,8]

model=Sequential()
model.add(Dense(12,input_dim=8,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(x,y,epochs=2000,batch_size=10)

scores=model.evaluate(x,y)
print('\n%s: %.2f%%'% (model.metrics_names[1],scores[1]*100))
```
